```python
from collections import defaultdict
summaryDictionary = defaultdict(int)
detailedDictionary = defaultdict(int)
data = pd.read_csv('gdrive/My Drive/Colab Notebooks/MergedUserStudy2.csv')
data.head(10)
for i in range(0,2400,12):
  resolution = data['Resolution'][i]
  up = 0
  down = 0 
  right = 0
  left = 0
  for z in range(12):
    print(data['Shape Drawn'][i+z])
    if(data['Shape Drawn'][i+z] == data['Shape Percieved'][i+z]):
      if(data['Shape Drawn'][i+z] == 'up'):
        up = up + 1
      elif(data['Shape Drawn'][i+z] == 'down'):
        down = down + 1
      elif(data['Shape Drawn'][i+z] == 'left'):
        left = left + 1
      elif(data['Shape Drawn'][i+z] == 'right'):
        right = right + 1
      print(up, left , right , down)
  print("----------")

  if(up >= 2):
    summaryDictionary[resolution] += 1
    detailedDictionary[str(resolution)+"up"] += 1
  if(down >= 2):
    summaryDictionary[resolution] += 1
    detailedDictionary[str(resolution)+"down"] += 1
  if(left >= 2):
    summaryDictionary[resolution] += 1
    detailedDictionary[str(resolution)+"left"] += 1
  if(right >= 2):
    summaryDictionary[resolution] += 1
    detailedDictionary[str(resolution)+"right"] += 1
  print(summaryDictionary)
  print(detailedDictionary)
  print("----------")


```
